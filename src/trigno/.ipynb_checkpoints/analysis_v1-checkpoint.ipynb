# EMG Signal Analysis

## Filtering and Smoothing

```python
muscle_emg_filtered, muscle_emg_rectified, muscle_emg_smoothed, muscle_emg_normalized = emg_filters(muscle_emg_raw, emg_time, muscle_name)
#plot_emg_signal(muscle_emg_normalized[0], emg_time[0], title=f"{muscle_name} EMG Normalized Signal #1", ylabel="%MVC")
```

## Windowing and Feature Extraction

```python
normalized_win, filtered_win, time_win = extract_emg_windows(muscle_emg_normalized, muscle_emg_filtered, emg_time)
```

```python
feature_list_norm = ['MAV', 'WL', 'RMS', 'VAR', 'IAV', 'KURT', 'SKEW', 'AR4', 'LD', 'SAMPEN', 'MPK']
feature_list_filt = ['ZC', 'SSC', 'WAMP']

time_features_norm = extract_emg_features(normalized_win, feature_list_norm)
time_features_filt = extract_emg_features(filtered_win, feature_list_filt)
```

```python
dfs = []

for feature_dict, filt_dict in zip(time_features_norm, time_features_filt):
    # Flatten dei dizionari (trasforma ogni array 2D in 1D)
    flat_dict = {k: np.round(np.array(v).ravel(), 10) for k, v in feature_dict.items()}
    flat_filt = {k: np.round(np.array(v).ravel(), 10) for k, v in filt_dict.items()}

    
    # Combina le due serie di feature
    all_features = {**flat_dict, **flat_filt}
    
    # Crea DataFrame e aggiungi l'indice di finestra
    features_df = pd.DataFrame(all_features)
    features_df.insert(0, 'window_idx', range(len(features_df)))
    
    dfs.append(features_df)

# Concatenazione finale
dfs = pd.concat(dfs, ignore_index=True)
print(dfs)
```

## Label Assignment

```python
import ruptures as rpt
import matplotlib.pyplot as plt

def filter_breakpoints_min_distance(bkps, min_distance):
    if not bkps:
        return []
    filtered = [bkps[0]]
    for b in bkps[1:]:
        if b - filtered[-1] >= min_distance:
            filtered.append(b)
    return filtered

all_bkps_list = []

for idx, Smoothed_EMG in enumerate(muscle_emg_normalized):
    threshold = 0.02
    #0.05 heavy
    # Trova l'ultimo indice in cui il segnale è sopra la soglia
    valid_indices = np.where(Smoothed_EMG > threshold)[0]

    if len(valid_indices) > 0:
        n_samples = valid_indices[-1] + 1  # Include l'ultimo indice valido
    else:
        n_samples = 0  # tutto il segnale è sotto soglia
    
    one_third = n_samples // 3
    two_third = 2 * n_samples // 3

    # Primo segmento
    segment1 = Smoothed_EMG[:one_third]
    bkps1 = rpt.Binseg(model="l2").fit(segment1).predict(n_bkps=5)[:-1]
    left_bkp1 = min(bkps1, key=lambda x: x)  # più a sinistra
    right_bkp1 = max(bkps1, key=lambda x: x)  # più a destra

    # Secondo segmento
    segment2 = Smoothed_EMG[one_third:two_third]
    bkps2 = rpt.Binseg(model="l2").fit(segment2).predict(n_bkps=5)[:-1]
    bkps2 = [b + one_third for b in bkps2]
    left_bkp2 = min(bkps2, key=lambda x: x)
    right_bkp2 = max(bkps2, key=lambda x: x)

    # Terzo segmento
    segment3 = Smoothed_EMG[two_third:]
    bkps3 = rpt.Binseg(model="l2").fit(segment3).predict(n_bkps=5)[:-1]
    bkps3 = [b + two_third for b in bkps3]
    left_bkp3 = min(bkps3, key=lambda x: x)
    right_bkp3 = max(bkps3, key=lambda x: x)

    # Unione ordinata dei breakpoint
    all_bkps = sorted([left_bkp1, right_bkp1, left_bkp2, right_bkp2, left_bkp3, right_bkp3])
    all_bkps_list.append(all_bkps)
    
    # Plot
    plt.figure(figsize=(12, 5))
    plt.plot(Smoothed_EMG, label=f"Normalized EMG Signal {idx+1}")
    for i, cp in enumerate(all_bkps):
        plt.axvline(x=cp, color='red', linestyle='--', label='Change point' if i == 0 else "")
    plt.title(f"Signal {idx+1} - Change Point Detection")
    plt.xlabel("Sample Index")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.show()
```

```python
def assign_labels_by_indices(total_windows, start_indices, labels):
    assigned_labels = [''] * total_windows

    for i in range(len(start_indices)):
        start = start_indices[i]
        end = start_indices[i+1] if i+1 < len(start_indices) else total_windows

        for w in range(start, end):
            assigned_labels[w] = labels[i]

    return assigned_labels
```

```python
#labels = ['no weight', 'light', 'no weight', 'light', 'no weight', 'light', 'no weight']
labels = ['no weight', 'heavy', 'no weight', 'heavy', 'no weight', 'heavy', 'no weight']
dt = np.mean(np.diff(emg_time[0]))   
fs = 1 / dt                    

# Parametri finestra
window_duration = 0.2  # secondi
overlap = 0.5          # 50% overlap

window_size = int(window_duration * fs)
step = int(window_size * (1 - overlap))

assigned_labels_list = []  # lista dove mettere tutte le label assegnate

for i, all_bkps in enumerate(all_bkps_list):
    total_windows = len(normalized_win[i])# lunghezza del DataFrame corrispondente
    windows_start = [0] + all_bkps[:]
    windows_start_windows = [w for idx in windows_start for w in [idx // step] if w < total_windows]
    
    assigned_labels = assign_labels_by_indices(total_windows, windows_start_windows, labels)
    assigned_labels_list.append(assigned_labels)

all_assigned_labels = [label for sublist in assigned_labels_list for label in sublist]
```

```python
dfs['label'] = all_assigned_labels
```
